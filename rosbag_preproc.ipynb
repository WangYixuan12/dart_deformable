```python
import rosbag
import time
import cv2
```

```python
def bag_to_img(bag_path, dir_path):
    bag = rosbag.Bag(bag_path)
    for topic, msg, t in bag.read_messages(topics=['chatter', 'numbers']):
        print(msg)
    bag.close()
```

```python
bag_path = '/home/yixuan/Downloads/rope_edge_cover_3.bag'
bag = rosbag.Bag(bag_path)
```

```python
help(bag)
```

```python
from cv_bridge import CvBridge
bridge = CvBridge()
```

```python
topic_list = ['image_depth_rect', 'image_color_rect']
dir_path = '/home/yixuan/dart_deformable/data/rope_edge_cover_3/'
import os
os.makedirs(dir_path)
for topic in topic_list:
    os.makedirs(dir_path+topic+'/')
i = 0
for topic, msg, t in bag.read_messages(topics=topic_list):
    if topic == 'image_depth_rect' or topic == 'image_color_rect':
        cv_image = bridge.imgmsg_to_cv2(msg, desired_encoding='passthrough')
        cv2.imwrite(dir_path+topic+'/'+topic+'_'+str(i)+'.jpg', cv_image)
    i += 1
```
